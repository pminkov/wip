```python
import matplotlib.pyplot as plt

x = [5]
y = [5]
c = [10]

import random

for ii in range(0, 500):
  x.append(x[-1] + random.uniform(-2, 2))
  y.append(y[-1] + random.uniform(-2, 2))
  c.append(x[-1] + y[-1])

plt.plot(x, color='r')
plt.plot(y, color='g')
plt.plot(c, color='b')
plt.show()
```

```python
a = []
N = 100
for i in range(0, N):
    a.append([5])
import random

for ii in range(0, 500):
    for j in range(0, N):
        a[j].append(a[j][-1] + random.uniform(-2, 2))

for j in range(0, N):
    plt.plot(a[j], color='g')
plt.show()
```
